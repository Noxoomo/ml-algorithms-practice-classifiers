```python
def kappa(real, predicted):
    counts = np.zeros((3, 3))
    for i in range(real.size):
        counts[real[i], predicted[i]] += 1
    total = counts.sum()
    pra = counts[0, 0] + counts[1, 1] + counts[2, 2]
    pra = pra / total
    pre = (counts.sum(0) / total) * (counts.sum(1) / total)
    pre = pre.sum()
    return (pra - pre) / (1 - pre)
```
